```python
# Basic Line Plot
"""
 Key Features Used:

label → Adds a label for the legend
color, linestyle, linewidth → Customize appearance
xlabel(), ylabel() → Label axes
title() → Adds a title
grid(True) → Shows grid
legend() → Displays the legend Key Features Used:

"""

#Line plots (plt.plot) → Show trends over time (e.g., stock prices, temperature changes).
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine Wave", color="blue", linestyle="--", linewidth=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#scatter plot:Scatter plots (plt.scatter) → Show correlations between two numeric variables.
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="red", marker="o", alpha=0.7)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values, color="green")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()



# Use barh() for a horizontal bar chart.
```

```python
#Bar charts (plt.bar) → Compare discrete categories (e.g., sales by product, population by country)
#autopct → Displays percentages
labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [30, 25, 35, 10]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=["red", "yellow", "pink", "brown"])
plt.title("Pie Chart Example")
plt.show()
```

```python
#Histograms (plt.hist) → Show how data is spread (e.g., exam scores, income distribution).
"""bins=30 → Controls number of bars
edgecolor → Outlines bars
alpha → Adjusts transparency
"""
data = np.random.randn(1000)  # Generate 1000 random numbers

plt.hist(data, bins=30, color="purple", edgecolor="black", alpha=0.7)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
#boxplot:Box plots (plt.boxplot) → Highlight outliers and percentiles.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 1000 random numbers
data = np.random.randn(1000)

# Create a boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(y=data, color="purple")  # Vertical boxplot
plt.ylabel("Values")
plt.title("Boxplot Example")
plt.show()
```
